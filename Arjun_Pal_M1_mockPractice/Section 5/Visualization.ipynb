```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Create Own Dataset
data = {
    "student": ["John", "Emma", "Liam", "Olivia", "Noah", "Ava", "Ethan", "Sophia"],
    "subject": ["Math", "Math", "Science", "Science", "English", "English", "Math", "Science"],
    "score": [78, 85, 92, 88, 67, 73, 80, 95]
}

df = pd.DataFrame(data)

print("Dataset:\n")
print(df)

# Step 2: Bar Chart
# Average score by subject
grouped = df.groupby("subject")["score"].mean()

plt.figure()
grouped.plot(kind="bar")
plt.title("Average Score by Subject")
plt.xlabel("Subject")
plt.ylabel("Average Score")
plt.show()

# Step 3: Line Chart
# Student scores trend
plt.figure()
df["score"].plot(kind="line")
plt.title("Student Scores Trend")
plt.xlabel("Index")
plt.ylabel("Score")
plt.show()

# Step 4: Probability
# Assume pass if score >= 75
passed = df[df["score"] >= 75]
total_students = len(df)

probability = len(passed) / total_students

print("\nProbability of passing:", probability)
```
