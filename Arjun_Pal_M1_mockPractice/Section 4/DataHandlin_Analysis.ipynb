```python
# Import pandas
import pandas as pd

# Step 1: Create Dataset (CSV)
data = {
    "id": [1,2,3,4,5,6,7,8,9],
    "name": ["Alice","Bob","Charlie","Alice","Bob","Charlie","Alice","Bob","Charlie"],
    "subject": ["Math","Math","Math","Science","Science","Science","English","English","English"],
    "score": [85,70,60,90,75,65,88,72,68],
    "grade": ["A","B","C","A","B","C","A","B","C"]
}

df = pd.DataFrame(data)

# Save to CSV
df.to_csv("students.csv", index=False)

print("Dataset created:\n")
print(df)

# Step 2: Load Dataset
df = pd.read_csv("students.csv")

print("\nLoaded Dataset:\n")
print(df)

# Step 3: Filtering
print("\nStudents with score > 75:\n")
filtered = df[df["score"] > 75]
print(filtered)

# Step 4: Grouping
print("\nAverage score by subject:\n")
grouped = df.groupby("subject")["score"].mean()
print(grouped)

# Step 5: Sorting
print("\nStudents sorted by score (highest first):\n")
sorted_df = df.sort_values(by="score", ascending=False)
print(sorted_df)

# Step 6: Insights

# Insight 1: Highest average subject
top_subject = grouped.idxmax()
print(f"\nInsight 1: Subject with highest average score is: {top_subject}")

# Insight 2: Top performing student
top_student = sorted_df.iloc[0]
print("\nInsight 2: Top performing student:")
print(top_student)
```
